# Eliminação Gaussiana

Para escalonar, podemos definir a _k-ésima_ linha por:

$L_k^{(i)}<-L_k^{(i-1)}-m_{kl}L_l^{(i-1)}$

onde $m_{kl}$ é o pivô definido a partir de:

$m_{kl} = \frac{A_{kl}}{A_{ll}}$

```python
def eliminacao_gaussiana(A):
    matriz_escalada = [linha.copy() for linha in A]

    for i in range(len(A)):
        for k in range(i+1, len(A)):
            pivot = matriz_escalada[k][i] / matriz_escalada[i][i]
            matriz_escalada[k] = [(matriz_escalada[k][x] - pivot * matriz_escalada[i][x]) for x in range(len(matriz_escalada[k]))]
    
    return matriz_escalada
```

```python
A = [
    [ 4,  3, -1],
    [ 2,  1,  1],
    [-1,  3,  4]
]

print(*eliminacao_gaussiana(A), sep="\n")
```

```python
A = [
    [1, 1],
    [2, 1]
]

print(*eliminacao_gaussiana(A), sep="\n")
```

```python
A = [
    [1, -2, 1],
    [0, 2, -1],
    [2, 3, 1]
]

print(*eliminacao_gaussiana(A), sep="\n")
```

```python
A = [
    [0, -2, 1],
    [1, 2, -1],
    [2, 3, 1]
]

try:
    print(*eliminacao_gaussiana(A), sep="\n")
except ZeroDivisionError:
    print("Erro: Divisão por zero")
```

```python
A = [
    [1e-6, -2, 1],
    [0, 2, -1],
    [2, 3, 1]
]

try:
    print(*eliminacao_gaussiana(A), sep="\n")
except ZeroDivisionError:
    print("Erro: Divisão por zero")
```
