## Escalonamento de matriz

```python
import numpy as np
```

```python
def solucao(matriz, vetor):
    print("========================")
    matriz_solu=[]
    for i in range(len(matriz)):
        pos_f=len(matriz)-i

        result=vetor[pos_f-1][0]
        for j in range(i):
            result-=matriz_solu[j]*matriz[pos_f-1][len(matriz)-1-j]

        result=result/matriz[pos_f-1][pos_f-1]
        print(f"O x{pos_f} = {result}")
        matriz_solu.append(result)
```

```python
def pivo(matriz, vetor, posi):
    if not matriz[posi][posi]:
        for i in range(posi, len(matriz)):
            if matriz[i][posi]!=0:
                matriz[posi], matriz[i] = matriz[i], matriz[posi]
                vetor[posi], vetor[i] = vetor[i], vetor[posi]
                return False

            if i == len(matriz)-1:
                return True
    
    return False
            
```

```python
def escalona(matriz, vetor_resp):
    for i in range(len(matriz)):
        for j in range(len(matriz)-1, i, -1):
            if i!=j:
                if pivo(matriz, vetor_resp, i):
                    continue

                m=matriz[j][i]/matriz[i][i]
                Lk=np.array(matriz[j])
                Lk1=np.array(matriz[i])
                matriz[j]=list(Lk-m*Lk1)

                Lk=np.array(vetor_resp[j])
                Lk1=np.array(vetor_resp[i])
                vetor_resp[j]=list(Lk-m*Lk1)
    
    for i in range(len(matriz)):
        for j in range(len(matriz)):
            matriz[i][j]=float(matriz[i][j])
            vetor_resp[i][0]=float(vetor_resp[i][0])

    
    return matriz, vetor_resp
```

```python
matriz, vetor=escalona(
[
     [2,4],
     [1,3]
], 


[
    [3],
    [7]
]
)

for i in range(len(matriz)):
    print(matriz[i], end=" = ")
    print(vetor[i])

solucao(matriz, vetor)
```
