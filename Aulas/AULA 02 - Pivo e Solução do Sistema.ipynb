## Escalonamento de matriz

```python
import numpy as np
```

```python
def escalona(matriz):
    linha_lim=0
    col_in=0

    while True:
        for i in range(len(matriz)-1,linha_lim,-1):
            linha=0
            while True:
                if matriz[i-1-linha][col_in]!=0:
                    break
                
                linha+=1

            if i-linha-1==0 and col_in!=0:
                continue

            m=matriz[i][col_in]/matriz[i-1-linha][col_in]
            Lk=np.array(matriz[i])
            Lk1=np.array(matriz[i-1-linha])
            matriz[i]=list(Lk-m*Lk1)
        
        if col_in+1==len(matriz):
            break

        col_in+=1
        linha_lim+=1
    
    for i in range(len(matriz)):
        for j in range(len(matriz)):
            matriz[i][j]=float(matriz[i][j])

    
    return matriz



for i in escalona([[1,2,3],
     [2,4,7],
     [2,4,9]]):
    print(i)
```

```python
def teste(func, dici):
    for i in range(1,len(dici.keys())+1):
        try:
            bol=(func(dici[i][0])==dici[i][1])
            print(f"Teste {i}: {bol}")
        except:
            print(f"Teste {i}: Erro")
    
teste(escalona, dicionario_teste)
```
